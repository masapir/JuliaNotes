```julia
using PyPlot
```

```julia
x = linspace(0, 5pi, 1000) # 1000 points from 0 to 2pi
y = 0.6sin(.8x) # stretch and shrink sine function
b = sin(x) # baseline sine function
c = y + x # push y up by x
t = b + x
plot(x, y, color="green", linewidth=0.9)
plot(x, b, color="red", linewidth=0.9)
plot(x, x, color="black", linewidth=0.7)
plot(x, c, color="blue", linewidth=0.9)
plot(x, t, color="brown", linewidth=1)
title("infinite number of fixed points")
xlabel("x")
```

```julia
x = linspace(0, 5, 1000) # 
y = x^2
plot(x, y, color="green", linewidth=0.9)
#plot(x, b, color="red", linewidth=0.9)
plot(x, x, color="black", linewidth=0.7)
title("circle")
xlabel("x")
```
