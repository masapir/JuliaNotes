**Short Circuit Evaluation**

&& (and) and || (or) are Boolean operators. 

 In **expression-a && expression-b**, expression-b is evaluated only if expression-a evaluated to true. If expression-a evaluated to false there is no need to evaluate expression-b because the whole && expression cannot be true.  expression-b is said to be short circuited when expression-a is false. 

Similarly in **expression-a || expression-b**, expression-b is evaluated only if expression-a evaluated to false.

Because of the way && and || expressions are evaluated, they can be used to make short hand conditional statements.

conditional-test && result-expression may be understood as: if conditional-test then result-expression.

Similarly  conditional-test || result-expression may be understood as if conditional-test is false then evaluate result-expression.

By the way & and | are then and and or Boolean operators without short circuit evaluation.

**Newton's Method To Compute The Square Root Of x**

This method tests whether a guess is good enough.  A guess is good enough if the square of the guess is within a specified tolerance of x.  Guess is returned if it is good enough.  Otherwise the procedure forms a new guess = average of x and x/guess.

```julia
# Average returns the arithmetic average of 2 numbers x and y.

Average(x, y) = (x + y)/2
```

```julia
Average(10, 20)
```

```julia
Average(0, 1)
```

```julia
Average(-1, 1)
```

```julia
# GoodEnough returns true if guess^2 is within tolerance ϵ of x, false otherwise

GoodEnough(guess, x, ϵ) = abs(guess^2 - x) < ϵ
    
```

```julia
GoodEnough(1.9, 4.0, .001)
```

```julia
GoodEnough(1.99, 4, 0.001)
```

```julia
GoodEnough(1.9999999999, 4, 0.001)
```

```julia

# Improve returns a better guess.
# Newton noticed that iteratively taking 
# the average of the guess and guess/x
# yielded better and better guesses to the √x

Improve(guess, x) = Average(guess, x/guess)

```

```julia
Improve(1.9, 4)
```

```julia
Improve(1.875, 4)
```

```julia
# NewtonIterate describes Newton's method to compute
# the √x by iteratively improving a guess (at the √x) to within a 
# specified tolerance.
# Note after the else statement, a recursive call with improved guess
# Note also this is another way to define a function.

function NewtonIterate(guess, x, ϵ)
    if GoodEnough(guess, x, ϵ)
            guess
        else
        NewtonIterate(Improve(guess, x), x, ϵ)
        end
end
```

```julia
NewtonIterate(1.0, 4, 0.001)
```

By the way NewtonIterate illustrates how a recursive process may be used to iterate over a procedure without having to use a loop syntax like for or while. An iterative process may also be implemented by maintaining computation state in a set of variables and incrementing and adjusting the state of those variables as a process is repeated until a final condition is reached.

```julia

# NewtonSqrt computes the √x to within a specified tolerance ϵ
# given x and ϵ (optional) using NewtonIterate called 
# with an initial guess of 1.0.

NewtonSqrt(x, ϵ=0.0001) = NewtonIterate(1.0, x, ϵ)
```

The next function, SmallestDivisor, returns the smallest divisor of a number, x. This is accomplished by iterating through integers from 2 through the biggest integer i such that i^2 not greater than x because if there is no divisor less than i then the smallest divisor of x is x.

```julia
function SmallestDivisor(x)
    function FindDivisor(x, denom)
        if denom^2 > x
            x
        elseif IsDivider(denom, x)
            denom
        else FindDivisor(x, denom+1)
        end
    end
    FindDivisor(x, 2) # start FindDivisor with integer 2
end
```

**Fix the Fermat Test below**

**Fermat's Little Theorem**  If p is a prime number and a is a positive integer less than p, then a^p is congruent to a modulo p.  That is:  the remainder of a^p/p = the remainder of a/p.  Now let's look at a probabilistic test for primality for an integer n.

Pick an integer a < n.  Is the remainder of a^n/n = remainder of a/n?  If yes, then n might be a prime number.  If no, then n is certainly not a prime number.  The more integers a < n that test "yes" the more certainty we have that n is prime.

**Carmichael Numbers**  Some numbers will keep testing "yes--might be prime" even though they are not prime. They trick this algorithm.

```julia
function Fermat(number::BigInt)
    function FermatTest(base::BigInt, number::BigInt)
        if base == number
            true
        elseif (base^number)%number == base%number
            FermatTest(base+1, number)
        else false
        end
    end
    FermatTest(2, number)
end
            
```

**Find A Few Carmichael Numbers** So let's find the Carmichael numbers up to 10,000. These are the number that are not prime even though the Fermat test suggests they are prime. That is, Carmichael numbers look prime from the Fermat test but they are not prime.

```julia
function Carmichael(cutoff=10000) # specify a cutoff or default to 10,000
    Carmichaels = []
    function CarmichaelTest(cutoff, candidate)
        if candidate == cutoff
            println("Carmichael numbers up to ", cutoff, ": ", Carmichaels)
        elseif Fermat(candidate) & !(IsPrime(candidate))
            cat(1, Carmichaels, [candidate])
        else CarmichaelTest(cutoff, candidate + 1)
        end
    end
    CarmichaelTest(cutoff, 1)
end
                
```

Series also illustrate a pattern than can be generalized.

The pattern consists in summing terms from a starting point a through an ending point b and where the next term is always derived from a term by applying some single transformation (function next) to the term.

**∑ summation** of terms from a to b with a general next function:

```julia
function ∑(term, a, next, b) # args term and next are procedures; a and b are integers
    if a > b
        0
    else
        term(a) + ∑(term, next(a), next, b)
    end
end
```

**Simpson Rule Approximation To Compute An Integral**.   The integral of f over a to b is approximately:   h/3  *  [y0 + 4y1 + 2y2 + 4y3 + 2y4 + 4y5 + . . . + yn] with n even and h = (b - a)/n, yk = f(a + kh)


```julia
function simpson(f, a, b, n)    # n must be even; integrating over [a,b]
    sNext(x) = x + (b - a)/n   # to get next x value, add increment
    function sTerm(x)          # function to return the term for the sum
        if x==a  # if x is an end point, term is f(x)
            f(x)
        elseif x==b
            f(x)
        elseif (n*(x - a)/(b - a))%2 != 0    # if k is odd
            4*f(x)                               # term is 4f(x)
        else
            2*f(x)                          # otherwise term is 2f(x)
        end
    end
    ((b - a)/(3*n))*∑(sTerm, a, sNext, b)
end
```

**Iterative Summation Process** Let's write a version of the summation function that is iterative rather than recursive by keeping track of state variables.

```julia
function IterativeΣ(a, term, next, b)
    function IterateToSum(a, result)
        if a > b
            result
            else IterateToSum(next(a), result + term(a))
        end
    end
    IterateToSum(a, 0)
end
```

The next function, accumulate, generalizes the pattern described in the summation and product functions.

```julia
function accumulate(combiner, null_value, func, a, next, b)
    if a > b
        null_value
    else
        combiner(func(a), accumulate(combiner, null_value, func, next(a), next, b))
    end
end
```

**Filtered Accumulate**. Accumulates only function values for elements in domain that pass a specified filter.

```julia
function Filtered_Accumulate(filt, combiner, null_value, func, a, next, b)
    if a > b
        null_value
        elseif filt(a) # if x passes the filter then combine f(x) with combination:
        combiner(func(a), Filtered_Accumulate(filt, combiner, null_value, func, next(a), next, b))
    else
        Filtered_Accumulate(filt, combiner, null_value, func, next(a), next, b)
    end
end
```
